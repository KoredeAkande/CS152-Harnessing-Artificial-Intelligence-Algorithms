```python
### NUMBERED MENU-BASED RESPONSE (YOU CAN ALSO INPUT THE ACTUAL OPTION) #####################
import tempfile
import os
from git import Repo
import re
from time import sleep


# Check if pyswip package needs to be installed

try:
    from pyswip.prolog import Prolog
    from pyswip.easy import *
except ModuleNotFoundError as err:
    os.mkdir("PySWIP")
    Repo.clone_from("https://github.com/rohanshekhar/pyswip.git", "PySWIP/")
    os.symlink("PySWIP/pyswip/", "pyswip")
    from pyswip.prolog import Prolog
    from pyswip.easy import *

prolog = Prolog()  # Global handle to interpreter

retractall = Functor("retractall")
known = Functor("known", 3)


# Define foreign functions for getting user input and writing to the screen
def write_py(X):
    sys.stdout.flush()
    print(str(X))
    return True

#Function to print options as a numbered menu, where options can be chosen through number or text
def menu(opt_list):
    
    #Iterate through options
    for num in range(len(opt_list)):
        #Number options
        if '_' in opt_list[num]:
            print(str(num+1) + ')',opt_list[num].replace("_"," ").capitalize())
        else:
            print(str(num+1) + ')',opt_list[num].capitalize())
            
    print('----------------------')

def read_py(A, V, Y, options):

    str_opt= re.findall(r'\w+|\$+',str(options))
    
    if isinstance(Y, Variable):
        
        #changing text based on the question
        if str(A) == 'price':
            print("What's your preferred " + str(A)+ " point?")
            
        elif str(A) == 'city':
            print("What rotation city are you based in? ")
            
        elif str(A) == 'distance':
            print("How far from the res hall are you looking to go to? ")
            
        elif str(A) == 'place':
            print("What kind of " + str(A) + " would you like to work at? ")
            
        elif str(A) == 'ratings':
            print("Do " + str(A) +" matter for you or not really? ")
            
        elif str(A) == 'wifi':
            print("Do you need a location with good " + str(A) +"?")
            
        elif 'symptom' in str(A):
            print("Are you showing any of the symptoms below? (input if applicable)")
            
        elif str(A) == 'unwell':
            print("Are you feeling unwell?")
            
        elif str(A) == 'covid_contact':
            print("Have you come in contact with anyone who has tested positive or who was showing covid symptoms?")
            
        elif str(A) == 'insurance':
            print("Do you want to find a hospital inside Minerva's health plan network of hospitals?")
            
        else:
            print("What's your preferred " + str(A)+ " type?")
            
        menu(str_opt)
        response = input('Enter your choice:')
        
        # check if multiple inputs (covid symptoms)
        if ',' in response:
            response = response.split(",")
            
            for number in response:
                number = number.strip()
                # raise exception if input is not a digit
                if not number.isdigit():
                    raise Exception("Your input needs to be one of the digits in the options above.")
                    
                try:
                    Y.unify(str_opt[int(number)-1])
                    return True
                except ValueError:
                    Y.unify(number)
                    return True
        else:
        
            # raise exception if input is not a digit
            if not response.isdigit():
                raise Exception("Your input needs to be one of the digits in the options above.")

            try:
                Y.unify(str_opt[int(response)-1])
                return True
            except ValueError:
                Y.unify(response)
                return True
            
    return False

def covid_expert():
    write_py.arity = 1
    read_py.arity = 4

    registerForeign(read_py)
    registerForeign(write_py)

    # Create a temporary file with the KB in it
    (FD, name) = tempfile.mkstemp(suffix='.pl', text="True")
    with os.fdopen(FD, "w") as text_file:
        text_file.write(KB)
    prolog.consult(name)  # open the KB for consulting
    os.unlink(name)  # Remove the temporary file

    call(retractall(known))
    # we are querying study place >>

    system_tasks= ['Covid-Safe Restaurants','Covid-Friendly Study Places','Covid Self-Checker & Other Health Options']

    print('--------------------------------------------------------------------')
    print("****************** COVID EXPERT SYSTEM *****************************")
    print('--------------------------------------------------------------------','\n')

    sleep(1)

    print('What would you like assistance with?')
    sleep(1)
    for task_num in range(len(system_tasks)):
        print(str(task_num+1) + ')',system_tasks[task_num])

    inputted_task= input('Your choice:')

    # raise exception if input is not a digit
    if not inputted_task.isdigit():
        raise Exception("Your input needs to be one of the digits in the options above (e.g. 1, 2, 3).")

    print('\n')
    sleep(0.5)

    if inputted_task == '1':
        suggestion = [s for s in prolog.query("eat(X).", maxresult=3)]
        print("We recommend " +
          " or ".join([suggestion[i]['X'] for i in range(len(suggestion))]) +
          "." if suggestion else "We didn't find a suitable restaurant based on your responses :(")

    elif inputted_task == '2':
        suggestion = [s for s in prolog.query("study(X).", maxresult=3)]
        print("We recommend " +
          " or ".join([suggestion[i]['X'] for i in range(len(suggestion))]) +
          "." if suggestion else "We didn't find a suitable study place based on your responses :(")

    else:
        print('What do you need?')
        options = ['Find Hospitals', 'COVID Self-Checker']
        for task_num in range(len(options)):
            print(str(task_num+1) + ')',options[task_num])

        hospital= input('Your choice:')

        if hospital == '1':
            suggestion = [s for s in prolog.query("health_assistance(X).", maxresult=3)]
            print(("We recommend " + (suggestion[0]['X'].replace('_',' ') + ".")) if suggestion else "We didn't found any hospital matching your criteria")
        else:
            suggestion = [s for s in prolog.query("advice(X).", maxresult=3)]
            print(("We recommend " + (suggestion[0]['X'].replace('_',' ') + ".")) if suggestion else "We don't think you have COVID, but be safe and take a bunch of precautions :(")
```
